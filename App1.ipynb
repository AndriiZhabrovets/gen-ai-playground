```python
import locale
import openai
import yaml
import os
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
import panel as pn
from IPython.display import display, Markdown,Latex,Image

pn.extension('katex',design="material", sizing_mode="stretch_width")
locale.getpreferredencoding = lambda: "UTF-8"

with open('chatgpt_api_credentials.yml', 'r') as file:
    api_creds = yaml.safe_load(file)
api_creds.keys()
os.environ['OPENAI_API_KEY'] = api_creds['openai_key']
openai.api_key  = api_creds['openai_key']
model= ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
#model = ChatOpenAI(model='gpt-4o',temperature=0)
```

# Use Prompt Templates and run LLM.

```python

SYS_PROMPT=f"""You are an expert in Math and an assistant creating high school problems.
            """
USER_PROMPT =f"""
             Generate a problem involving matrices and determinants.
             Choose an application in economics.
            """

messages = [
SystemMessage(content=SYS_PROMPT), 
HumanMessage(content=USER_PROMPT) 
]
response = model.invoke(messages)       
```
