```python
import requests

class WikiSearchResultsWithLLM:
    def __init__(self, search_term, language="en", results_limit=10, use_llm=False):
        self.search_term = search_term
        self.language = language
        self.results_limit = results_limit
        self.use_llm = use_llm
        self.base_url = f"https://{language}.wikipedia.org/w/api.php"
        
        # If using LLM, refine the search term first
        if self.use_llm:
            self.search_term = self._refine_search_term_with_llm(self.search_term)
        
        self.results = self._search_wikipedia()

    def _refine_search_term_with_llm(self, term):
        """Use an LLM to refine or enhance the search term."""
        
        messages = [
        SystemMessage(content=f"Refine the following search term for better Wikipedia results: {term}"), 
        HumanMessage(content="") # state["messages"])
        ]
        response = model.invoke(messages)

        refined_term = response.content
        return refined_term if refined_term else term

    def _search_wikipedia(self):
        """Private method to perform the search on Wikipedia."""
        params = {
            "action": "query",
            "list": "search",
            "srsearch": self.search_term,
            "format": "json",
            "srlimit": self.results_limit,
        }
        response = requests.get(self.base_url, params=params)
        data = response.json()

        # Parse search results
        search_results = data.get("query", {}).get("search", [])
        return search_results

    def get_results(self):
        """Return a structured list of search results with titles, snippets, and URLs."""
        return [
            {
                "title": result["title"],
                "snippet": result["snippet"],
                "url": f"https://{self.language}.wikipedia.org/wiki/{result['title'].replace(' ', '_')}"
            }
            for result in self.results
        ]
def Run_Wiki_search(search_term):
    print("search_term",search_term)
    search_tool = WikiSearchResultsWithLLM(search_term, results_limit=5, use_llm=False)
    results = search_tool.get_results()
    output=""  
    for result in results:
#    print(f"Title: {result['title']}\nSnippet: {result['snippet']}\nURL: {result['url']}\n")
        output=output + result['title']+ "\n" + result['snippet']+ "\n" +result['url']+"\n\n"
   
    return output
```
